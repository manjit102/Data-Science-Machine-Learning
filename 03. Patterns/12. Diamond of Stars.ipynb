# <font color = 'red'>Diamond of Stars</font>

Print the following pattern for the given number of rows.
* **Note : N is always odd**
* Pattern for N = 5
* ..@
* .@@@
* @@@@@
* .@@@
* ..@
* The dots represent spaces and @ represent the star.

```python
n = int(input('Enter the Number : '))

count = 1
r = ((n+1)//2)
while count <= r:
    i = 1
    while i <= (count+r-1):
        if i <= (r-count):
            print(' ', end = '')
        else:
            print('*', end = '')
        i = i + 1
    count = count + 1
    print()
count = 1
while count < r:
    i = 1
    while i <= n - count:
        if i <= count:
            print(' ', end = '')
        else:
            print('*', end = '')
        i = i+1
    count = count + 1
    print()
```
